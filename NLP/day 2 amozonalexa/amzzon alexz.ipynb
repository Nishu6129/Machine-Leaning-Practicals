```python
from textblob import TextBlob
```

```python
def get_sent(text):
    if isinstance(text, str):
        mytext=TextBlob(text)
        mysent=mytext.sentiment
        return mysent.polarity
    else:
        print(text)
        text=str(text)
        mytext=TextBlob(text)
        mysent=mytext.sentiment
        return mysent.polarity
```

```python
get_sent("Configuration: Fire TV Stick")
```

```python
dataset["polarity"]=dataset["verified_reviews"].apply(get_sent)
```

```python
dataset
```

```python
def get_sent_sub(text):
    if isinstance(text, str):
        mytext=TextBlob(text)
        mysent=mytext.sentiment
        return mysent.subjectivity
    else:
        print(text)
        text=str(text)
        mytext=TextBlob(text)
        mysent=mytext.sentiment
        return mysent.subjectivity
```

```python
dataset["subjectivity"]=dataset["verified_reviews"].apply(get_sent_sub)
```
